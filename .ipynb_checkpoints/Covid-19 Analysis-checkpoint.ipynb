# Importing Necessary lib

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import plotly.express as px
from datetime import datetime
```

# Loading dataset

```python
covid_df=pd.read_csv("https://raw.githubusercontent.com/abhisarahuja/Covid-19-Data-Analysis-Project-Using-Python-And-Tableau/main/covid_19_india.csv")
```

```python
covid_df.head()
```

```python
covid_df.tail()
```

```python
covid_df.dropna(inplace=True)
```

```python
covid_df.isnull()
```

```python
covid_df.drop(["Sno"],axis=1,inplace=True)
```

```python
covid_df.drop(["Time","ConfirmedIndianNational","ConfirmedForeignNational"],axis=1,inplace=True)
```

```python
covid_df.tail()
```

```python
covid_df["Date"]=pd.to_datetime(covid_df["Date"],format='%Y-%m-%d')
covid_df.tail()
```

```python
#Active cases
covid_df['Active_cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.tail()
```

# Staewise Info

```python
statewise=pd.pivot_table(covid_df,values=['Confirmed','Cured','Deaths'],index='State/UnionTerritory',aggfunc=max)
```

```python
statewise['Recovery Rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap='RdPu')
```

```python
#Top 10 states with active cases


top_states=covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
ax=sns.barplot(x='State/UnionTerritory',y='Active_cases',data=top_states.iloc[:10])
plt.title("Top 10 states with active cases.",size=20)
plt.xlabel('States/UnionTerritory')
plt.ylabel('Total active cases')
plt.show()
```

```python
#Top 10 states with max no of Deaths

top_states=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(15,5))
ax=sns.barplot(x='State/UnionTerritory',y='Deaths',data=top_states.iloc[:10],palette='bright')
plt.title("Top 10 states with max Deaths.",size=20)
plt.xlabel('States/UnionTerritory')
plt.ylabel('Total Death cases')
plt.show()
```

```python
#Growth Trends
fig=plt.figure(figsize=(15,7))
ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Delhi','Tamil Nadu','Uttar Pradesh'])], x='Date',y='Active_cases',hue='State/UnionTerritory')
ax.set_title("Top 5 states Growth Pattern",size=20)
```

```python
#Male vs Female
Male=vaccine_df['Male(Individuals Vaccinated)'].sum()
Female=vaccine_df['Female(Individuals Vaccinated)'].sum()
fig = px.pie(names=['Male','Female'],values=[Male,Female],title="Male  vs Female",color_discrete_sequence= ['#ADD8E6', '#FFC0CB']  )

fig.show()
```

```python
#remove ros where state=India
vaccine=vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
vaccine
```

```python
#Most vaccinated states
max_vac=vaccine.groupby(by='State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values(by='Total',ascending=False)[:5]
max_vac.head()
```

```python
#plotting max vaccinated states
fig=plt.figure(figsize=(8,5))
ax=sns.barplot(x=max_vac.index,y=max_vac.Total,data=max_vac.iloc[:10],palette='bright',width=.5)
plt.title("Top 5 states with max Vaccinated Individuals.",size=20)
plt.xlabel('States')
plt.ylabel('Vaccinated')
plt.show()
```

```python
#Least vaccinated
least_vac=vaccine.groupby(by='State')['Total'].sum().to_frame('Total')
least_vac=max_vac.sort_values(by='Total')
least_vac.head()
```

```python
#plotting least vaccinated state
fig=plt.figure(figsize=(8,5))
ax=sns.barplot(x=least_vac.index,y=least_vac.Total,data=least_vac.iloc[:10],palette='pastel',width=.5)
plt.title("5 states with least Vaccinated Individuals.",size=20)
plt.xlabel('States')
plt.ylabel('Vaccinated')
plt.show()
```
